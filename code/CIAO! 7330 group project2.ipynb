```python
# name the chart
# load dataset
# set canvas size
# Add a legend

plt.figure(figsize=(10,5))
plt.title('Financial Assistance Schemes for Primary & Secondary Students')
plt.xlabel('x-year',fontsize=14)
plt.ylabel('y-amount',fontsize=14)
plt.bar(df_subsidy['School Year'],df_subsidy['Full-year Subsidy Rate (Full)'],alpha = 0.5,facecolor = 'burlywood', lw=1, label='Full-year Subsidy Rate (Full)')
plt.bar(df_subsidy['School Year'],df_subsidy['Average Amount of Subsidy Disbursed'],alpha = 1,facecolor = 'bisque',lw=1,label='Average Amount of Subsidy Disbursed')
plt.bar(df_subsidy['School Year'],df_subsidy['Full-year Subsidy Rate (Half)'],alpha = 0.5, facecolor = 'darkorange',  lw=1, label='Full-year Subsidy Rate (Half)')
plt.legend(loc=2)
plt.show()
```
